```python
import requests
import pandas as pd
from io import StringIO, BytesIO

BASE_STATION = "https://www.waterqualitydata.us/data/Station/search"
BASE_RESULT  = "https://www.waterqualitydata.us/data/Result/search"
BASE_SUMMARY = "https://www.waterqualitydata.us/data/summary/monitoringLocation/search"
```
